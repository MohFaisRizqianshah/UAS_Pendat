### Identifikasi Outlier

outlier merupakan nilai yang jauh dari nilai-nilai yang umum atau pola yang ada dalam data. Kehadiran outlier dapat mempengaruhi analisis statistik dan machine learning karena dapat menyebabkan estimasi yang tidak akurat atau bias dalam model yang dibangun.

Karakteristik Outlier:
1. Outlier merupakan nilai yang berada jauh dari sebagian besar data dalam dataset.
2. Outlier tidak mengikuti pola atau distribusi umum dari sebagian besar data.
3. Outlier bisa berupa nilai yang sangat besar (nilai tinggi) atau sangat kecil (nilai rendah) dibandingkan dengan nilai-nilai lain dalam dataset.

```python
import pandas as pd
import numpy as np

# Fungsi untuk mencari outlier menggunakan pendekatan IQR
def find_outliers_iqr(data):
    # Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    
    # Menghitung jangkauan interkuartil (IQR)
    IQR = Q3 - Q1
    
    # Menentukan batas atas dan batas bawah untuk outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Mengidentifikasi outlier
    outliers = (data < lower_bound) | (data > upper_bound)
    
    return outliers

# Mencari outlier dalam dataset untuk setiap fitur numerik
for column in df_abalone.select_dtypes(include=['int64', 'float64']).columns:
    outliers = find_outliers_iqr(df_abalone[column])
    num_outliers = outliers.sum()  # Menghitung jumlah outlier
    
    print(f"\nOutlier(s) dalam kolom '{column}':")
    print(df_abalone[column][outliers])
    print(f"Total outlier dalam kolom '{column}': {num_outliers}")
```

# KNN

Metode KNN (K-Nearest Neighbors) adalah salah satu algoritma yang digunakan dalam pembelajaran mesin untuk klasifikasi dan regresi. Pendekatan ini termasuk dalam kategori pembelajaran berbasis instansi atau non-parametrik, yang berarti algoritma tidak memerlukan pembentukan model yang kompleks sebelumnya. Sebaliknya, algoritma KNN mengandalkan kumpulan data pelatihan yang telah diberi label untuk membuat prediksi tentang label atau nilai target dari data yang tidak terlihat.

Cara kerja algoritma KNN adalah dengan mencari sejumlah tetangga terdekat (nearest neighbors) dari titik data yang akan diprediksi. Tetangga terdekat ini dipilih berdasarkan kesamaan atau jarak antara fitur-fitur data. Metode KNN kemudian menggunakan mayoritas dari label-label tetangga terdekat (untuk klasifikasi) atau rata-rata dari nilai target tetangga terdekat (untuk regresi) sebagai prediksi untuk titik data yang tidak diketahui.

Proses algoritma KNN dapat dijelaskan dalam beberapa tahap:

1. Menentukan parameter K: Parameter K adalah jumlah tetangga terdekat yang akan digunakan dalam proses prediksi. Nilai K harus ditentukan sebelumnya oleh pengguna.

2. Menghitung Jarak: Algoritma menghitung jarak antara titik data yang akan diprediksi dengan setiap titik data dalam dataset pelatihan. Jarak ini biasanya dihitung menggunakan metrik jarak seperti Euclidean distance, Manhattan distance, atau Minkowski distance.

3. Memilih Tetangga Terdekat: Algoritma memilih K titik data dengan jarak terdekat dengan titik data yang akan diprediksi.

4. Menghitung Prediksi: Untuk masalah klasifikasi, algoritma mengambil mayoritas dari label-label tetangga terdekat sebagai prediksi kelas untuk titik data yang tidak diketahui. Untuk regresi, algoritma mengambil rata-rata dari nilai target tetangga terdekat sebagai prediksi.

Algoritma KNN memiliki beberapa kelebihan, seperti sederhana dalam implementasi dan dapat menangani data yang kompleks. Namun, algoritma ini juga memiliki beberapa kelemahan, termasuk sensitif terhadap skala dan dimensi data, serta memerlukan penyimpanan besar untuk dataset besar. Selain itu, KNN cenderung memiliki kinerja yang lebih buruk dibandingkan dengan algoritma lainnya dalam situasi di mana fitur-fitur data tidak memiliki relevansi yang jelas atau ketika ada noise dalam data.

### Mencari akurasi data menggunakan metode KNN

pada point ini, kita akan mencari akurasi dataset abalone sebelum di modifikasi data outliernya.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

#memasukkan data csv
abalone = pd.read_csv("Abalone.csv")
abalone

#membaca 5 data teratas
abalone.head()
abalone

#melihat tipe data yang akan digunakan
abalone.info()
abalone

#menentukan variabel independen
x = abalone.drop(["Sex"], axis=1)
x.head()

#menentukan variabel dependen
y = abalone["Sex"]
y.head()

#membagi data training dan data testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

#mengubah skala data
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

#mengaktifkan fungsi klasifikasi untuk knn
knn = KNeighborsClassifier (n_neighbors=3)

#memasukkan data training pada fungsi klasifikasi untuk knn
knn.fit(x_train, y_train)
```

```python
#menentukan prediksi
y_pred = knn.predict (x_test)

y_pred

#menentukan probabilitas prediksi
knn.predict_proba(x_test)

#menampilkan matriks hasil prediksi
print(confusion_matrix(y_test, y_pred))

#ketepatan hasil prediksi
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

#memasukkan data csv
abalone = pd.read_csv("outlier_abalone.csv")
abalone

#membaca 5 data teratas
abalone.head()
abalone

#melihat tipe data yang akan digunakan
abalone.info()
abalone

#menentukan variabel independen
x = abalone.drop(["Sex"], axis=1)
x.head()

#menentukan variabel dependen
y = abalone["Sex"]
y.head()

#membagi data training dan data testing
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

#mengubah skala data
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

#mengaktifkan fungsi klasifikasi untuk knn
knn = KNeighborsClassifier (n_neighbors=3)

#memasukkan data training pada fungsi klasifikasi untuk knn
knn.fit(X_train, y_train)

#menentukan prediksi
y_pred = knn.predict (X_test)
y_pred

#menentukan probabilitas prediksi
knn.predict_proba(X_test)

#menampilkan matriks hasil prediksi
print(confusion_matrix(y_test, y_pred))

#ketepatan hasil prediksi
print(classification_report(y_test, y_pred))
```

## Bagging (Bootstrap Aggregating)

Bagging adalah teknik yang melibatkan pembuatan beberapa subset dari data pelatihan dengan penggantian (metode bootstrap), kemudian melatih model yang berbeda pada setiap subset tersebut. Setelah itu, hasil prediksi dari masing-masing model ini digabungkan; untuk tugas regresi, hasil prediksi diambil rata-ratanya, sedangkan untuk tugas klasifikasi, hasil prediksi diambil berdasarkan mayoritas suara. Metode ini membantu mengurangi variabilitas dan meningkatkan akurasi prediksi.

<img src="image-20240612-143258.png" width="" align="" />

```python
import numpy as np
import matplotlib.pyplot as plt  # Added
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
import warnings

# Suppress the DataConversionWarning
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)

# Split the dataset into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

akurasi = []
n_est = []

# Loop over n_estimators from 3 to 20
for n_estimator in range(3, 21):
    # Initialize and fit the BaggingClassifier with GaussianNB as the base estimator
    clf = BaggingClassifier(base_estimator=GaussianNB(), n_estimators=n_estimator, random_state=0)
    clf.fit(X_train, y_train)
    
    # Predict the labels for the test set
    y_pred = clf.predict(X_test)
    
    # Calculate and print the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    akurasi.append(round(accuracy,2))
    n_est.append(n_estimator)
    print(f"n_estimators: {n_estimator}, Accuracy: {accuracy:.2f}")

# Plotting the results
plt.plot(n_est, akurasi, marker='o')
plt.title('Accuracy vs. Number of Estimators')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

Program ini menghasilkan berbagai nilai akurasi untuk n_estimators dari 3 hingga 20.
Hasil akurasi berkisar antara 0.49 dan 0.51, dengan nilai akurasi tidak banyak berubah seiring bertambahnya jumlah estimator.
Grafik yang dihasilkan menunjukkan hubungan antara jumlah estimator dan akurasi.
